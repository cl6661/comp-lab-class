```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np
import os
data=np.loadtxt("diff-wat.xvg",skiprows=20)
Nadata=np.loadtxt("diff-na.xvg",skiprows=20)
Cldata=np.loadtxt("diff-cl.xvg",skiprows=20)
print(data)
```

```python
step=data[:,0]
wat=data[:,1]
Na=Nadata[:,1]
Cl=Cldata[:,1]
plt.plot(step,wat,label="Water")
plt.plot(step,Na,label="Na")
plt.plot(step,Cl,label="Cl")
plt.legend(loc="upper right")
plt.legend(loc="upper right")
plt.xlabel('Time (ps)')
plt.ylabel('MSD (nm^2)')
plt.xlim([0,10000])
plt.title('Mean Square Displacement')
plt.savefig('Mean Square Displacement.pdf',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()
```
