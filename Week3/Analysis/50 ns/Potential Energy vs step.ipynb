```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np
import os
#f = open("potential.xvg", "r")
#print(f.read())
data=np.loadtxt("potential.xvg",skiprows=24)
#print(data)
step=data[:,0]
potential=data[:,1]
plt.plot(step,potential)
plt.xlabel('Energy Minimization Step')
plt.ylabel('Potential Energy (kJ/mol)')
plt.title('Energy Minimization')
plt.xlim([0,800])
plt.savefig('Potential_50ns.pdf',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()
```
