```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np
import os

data=np.loadtxt("pressure.xvg",skiprows=24)
#print(data)
time=data[:,0]
pressure=data[:,1]
plt.plot(time,pressure)
plt.xlabel('Time (ps)')
plt.ylabel('Pressure (bar)')
plt.title('Pressure')
plt.xlim([0,100])
plt.savefig('Pressure_50ns.pdf',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()
```
