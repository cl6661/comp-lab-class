```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np
import os
#f = open("gyrate.xvg", "r")
#print(f.read())
data=np.loadtxt("gyrate_50ns_nowater.xvg",skiprows=27)
#print(data)
step=data[:,0]/1000
radius=data[:,1]
plt.plot(step,radius)
plt.xlabel('Time (ns)')
plt.ylabel('Radius of gyration (nm)')
plt.title('Radius of gyration 1AKI_50ns_nowater, Unrestrained MD')
plt.xlim([0,50])
plt.ylim([1.3,1.5])
plt.savefig('Radius of gyration 1AKI_50ns_nowater.pdf',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()
```
