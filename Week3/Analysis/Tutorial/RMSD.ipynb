```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np
import os
#f = open("potential.xvg", "r")
#print(f.read())
data=np.loadtxt("rmsd.xvg",skiprows=18)
xtaldata=np.loadtxt("rmsd_xtal.xvg",skiprows=18)
#print(data)
step=data[:,0]
rmsd=data[:,1]
rmsdxtal=xtaldata[:,1]
plt.plot(step,rmsd,label="Ref: Equilibrated")
plt.plot(step,rmsdxtal,label="Ref: Crystal")
plt.legend(loc="upper right")
plt.legend(loc="upper right")
plt.xlabel('Time (ns)')
plt.ylabel('RMSD (nm)')
plt.title('RMSD 1AKI, Backbone')
plt.xlim([0,1])
plt.savefig('RMSD 1AKI_tutorial.pdf',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()
```
