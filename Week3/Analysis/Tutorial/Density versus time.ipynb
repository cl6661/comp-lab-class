```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np
import os

data=np.loadtxt("density.xvg",skiprows=24)
#print(data)
time=data[:,0]
density=data[:,1]
plt.plot(time,density)
plt.xlabel('Time (ps)')
plt.ylabel('Density (kg/m^3)')
plt.title('Density')
plt.xlim([0,100])
plt.savefig('Density_tutorial.pdf',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()
```
