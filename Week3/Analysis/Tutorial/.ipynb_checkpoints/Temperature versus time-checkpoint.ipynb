```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np
import os

data=np.loadtxt("temperature.xvg",skiprows=24)
#print(data)
time=data[:,0]
temperature=data[:,1]
plt.plot(time,temperature)
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.title('Temperature')
plt.xlim([0,100])
plt.savefig('Temperature.pdf',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()
```
