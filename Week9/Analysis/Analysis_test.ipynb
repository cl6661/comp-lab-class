```python
temp = [1.5, 1.0, 0.9, 0.8, 0.7, 0.65, 0.6, 0.55, 0.5, 0.475, 0.45]
for n in temp:
    # rename trajectory to traj with the correct temperature
    traj = "traj" + str(n).replace(".", "_")
    # link the trajectory to the variable traj
    traj = globals()[traj]
    # selecring the atoms 1 and 2 from the trajectories at each temperature
    traj_a1=traj.select_atoms("type 1")
    traj_a2=traj.select_atoms("type 2")
    # Calculating the average radial distribution function for atoms 1 to itself
    irdf11 = rdf.InterRDF(traj_a1, traj_a1, nbins=100, range=(0.0, 10.0))
    irdf11.run()
    plt.plot(irdf11.bins, irdf11.rdf)
    plt.xlabel('Radius (angstrom)')
    plt.ylabel('Radial distribution')
    plt.title('Radial distribution function for atoms 1 to itself at T = ' + str(n))
    plt.savefig('g_11_T' + str(n) + '.pdf', format='pdf',bbox_inches='tight',pad_inches=0.1)
    plt.show()
    # Calculating the average radial distribution function for atoms 2 to itself
    irdf22 = rdf.InterRDF(traj_a2, traj_a2, nbins=100, range=(0.0, 10.0))
    irdf22.run()
    plt.plot(irdf22.bins, irdf22.rdf)
    plt.xlabel('Radius (angstrom)')
    plt.ylabel('Radial distribution')
    plt.title('Radial distribution function for atoms 2 to itself at T = ' + str(n))
    plt.savefig('g_22_T' + str(n) + '.pdf', format='pdf',bbox_inches='tight',pad_inches=0.1)
    plt.show()
    # Calculating the average radial distribution function for atoms 1 to 2
    irdf12 = rdf.InterRDF(traj_a1, traj_a2, nbins=100, range=(0.0, 10.0))
    irdf12.run()
    plt.plot(irdf12.bins, irdf12.rdf)
    plt.xlabel('Radius (angstrom)')
    plt.ylabel('Radial distribution')
    plt.title('Radial distribution function for atoms 1 to 2 at T = ' + str(n))
    plt.savefig('g_12_T' + str(n) + '.pdf', format='pdf',bbox_inches='tight',pad_inches=0.1)
    plt.show()
    # Overlap of the radial distribution functions for atoms 1 to iself, atoms 2 to itself and atoms 1 to 2
    plt.plot(irdf11.bins, irdf11.rdf, label='g_11(r)')
    plt.plot(irdf22.bins, irdf22.rdf, label='g_22(r)')
    plt.plot(irdf12.bins, irdf12.rdf, label='g_12(r)')
    plt.xlabel('Radius (angstrom)')
    plt.ylabel('Radial distribution')
    plt.title('Radial distribution function for atoms 1 to itself, atoms 2 to itself and atoms 1 to 2 at T = ' + str(n))
    plt.legend()
    plt.savefig('g_11_22_12_T' + str(n) + '.pdf', format='pdf',bbox_inches='tight',pad_inches=0.1)
    plt.show()
```
