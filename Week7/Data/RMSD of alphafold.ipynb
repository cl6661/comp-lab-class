```python
# this notebook is for plotting
# the end to end analysis
import numpy as np
import matplotlib.pyplot as plt

#load mdtraj
import mdtraj as md

# load os to change python directory 
import os

# load the pdb for crystal
traj = md.load('1ubq.pdb')
print('crystal file:',traj)

# Slice CA in crystal file
crystal_CA = traj.topology.select("name CA")
traj_crystal_CA = traj.atom_slice(crystal_CA, inplace=False)
print('CA sliced crystal file:',traj_crystal_CA)

# Slice Backbond in crystal file
# crystal_bs = traj.topology.select("(backbone)")
crystal_bs = traj.topology.select("(backbone) or (sidechain)")
traj_crystal_bs = traj.atom_slice(crystal_bs, inplace=False)
print('BB and sc sliced crystal file:',traj_crystal_bs)
```
