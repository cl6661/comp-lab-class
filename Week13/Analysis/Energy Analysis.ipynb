# Plot of potential energy vs time for equilibrated MM system

```python
# read the .ener file in ../Outputs/NVT-1.ener directory
data=np.genfromtxt('../Outputs/NVT-1.ener',skip_footer=1)
# plot the data
plt.plot(data[:,0],data[:,4])
plt.xlabel('Time (ps)')
plt.ylabel('Potential Energy (kJ/mol)')
plt.title('Potential Energy vs. Time in NVT')
plt.savefig('../Figures/potential energy vs time in the NVT.pdf')
plt.show()

# read the .ener file in ../Outputs/NPT-1.ener directory
data=np.genfromtxt('../Outputs/NPT-1.ener',skip_footer=1)
# plot the data
plt.plot(data[:,0],data[:,4])
plt.xlabel('Time (ps)')
plt.ylabel('Potential Energy (kJ/mol)')
plt.title('Potential Energy vs. Time in NPT')
plt.savefig('../Figures/potential energy vs time in the NPT.pdf')
plt.show()
```
