```python
# this notebook is for plotting the reaction coordinate for the reaction versus time
import numpy as np
import pandas as pd
from pandas import read_csv
from matplotlib.image import NonUniformImage
import matplotlib.pyplot as plt
```

# Reaction Coordinate for the reaction versus time

```python
# Read Outputs/monitor/MONITOR-COLVAR.metadynLog
# loading COLVAR file
data = pd.read_csv('../Outputs/monitor/MONITOR-COLVAR.metadynLog', header=None, delim_whitespace=True,skiprows=0)
#plot coordinate vs time
plt.plot(data[0],data[1])
plt.xlabel('Time (fs)')
plt.ylabel('d1-d2 (Angstrom)')
plt.title('Reaction Coordinate versus time')
plt.savefig('../Figures/Reaction Coordinate versus time.pdf')
plt.show()
```

# QM/MM plus metadynamics

```python
# Read Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog
# loading COLVAR file
data = pd.read_csv('../Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog', header=None, delim_whitespace=True,skiprows=0)
#plot coordinate vs time
plt.plot(data[0],data[1])
plt.xlabel('Time (fs)')
plt.ylabel('d1-d2 (Angstrom)')
plt.title('Reaction Coordinate versus time after metad')
plt.text(0, -10,'Hill height: 1.5 kcal/mol \nHill width: 0.5 Angstrom\nHill freq: 100 steps')
plt.savefig('../Figures/Reaction Coordinate versus time after metad.pdf')
plt.show()
```
