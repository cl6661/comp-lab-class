```python
# this notebook is for plotting the end to end analysis and radius of gyration
import numpy as np
import matplotlib.pyplot as plt

#load mdtraj
import mdtraj as md

#read in the md trajectory
traj = md.load('1hz3_T310.stepid25000000.every100ps.nowater.xtc', top='1hz3_T310.start.nowater.gro')

# compute the distance between the first and last atom
def end_end(traj):
    # get the first and last atom
    first = traj.xyz[:,0,:]
    last = traj.xyz[:,-1,:]
    # compute the distance between the first and last atom
    dist = np.sqrt(np.sum((first-last)**2, axis=1))
    return dist

# compute the distance between the first and last atom
dist = end_end(traj)

# compute the radius of gyration
radius = md.compute_rg(traj, masses=None)

#plot distance vs time
plt.plot(traj.time, dist, label="End to End Distance")

#plot radius vs time
plt.plot(traj.time, radius, label="Radius of Gyration")
plt.legend(loc="upper right")
plt.xlabel('Time (ps)')
plt.ylabel('Distance (nm)')
plt.title('End to End Distance and Radius of Gyration')
plt.savefig('End to End Distance and Radius of Gyration',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()

#plot a normalized histogram of end to end distance
plt.hist(dist, histtype="step", density = True)
plt.xlabel('Distance (nm)')
plt.title('Normalized histogram of end-to-end distance')
plt.savefig('Normalized histogram of end-to-end distance',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()
plt.hist(radius, histtype="step", density = True)
plt.xlabel('Radius of gyration (nm)')
plt.title('Normalized histogram of Radius of gyration')
plt.savefig('Normalized histogram of Radius of gyration',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()
```
