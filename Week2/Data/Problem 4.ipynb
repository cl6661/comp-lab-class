```python
#read in the data file
newdata2=np.loadtxt("1hz3_T310.run.25000000.energy.xvg",skiprows=29)
time=newdata2[:,0]
potential=newdata2[:,1]
kinetic=newdata2[:,2]
totalenergy=newdata2[:,3]
temperature=newdata2[:,4]
pressure=newdata2[:,5]
volume=newdata2[:,6]

#change time to ns
timens=time/1000

#make 6 seperate plots
plt.plot(timens,potential)
plt.xlabel('Time (ns)')
plt.ylabel('Potential (kJ/mol)')
plt.title('Potential')
plt.savefig('Potential',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()
plt.plot(timens,kinetic)
plt.xlabel('Time (ns)')
plt.ylabel('Kinetic Energy (kJ/mol)')
plt.title('Kinetic Energy')
plt.savefig('Kinetic Energy',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()
plt.plot(timens,totalenergy)
plt.xlabel('Time (ns)')
plt.ylabel('Total Energy (kJ/mol)')
plt.title('Total Energy')
plt.savefig('Total Energy',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()
plt.plot(timens,temperature)
plt.xlabel('Time (ns)')
plt.ylabel('Temperature (K)')
plt.title('Temperature')
plt.savefig('Temperature',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()
plt.plot(timens,pressure)
plt.xlabel('Time (ns)')
plt.ylabel('Pressure (bar)')
plt.title('Pressure')
plt.savefig('Pressure',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()
plt.plot(timens,volume)
plt.xlabel('Time (ns)')
plt.ylabel('Volume (nm^3)')
plt.title('Volume')
plt.savefig('Volume',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()

#make a overlay plot of temperature and kinetic energy
x = timens
y1 = temperature
y2 = kinetic
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y1, 'g.')
ax2.plot(x, y2, 'b')

ax1.set_title('Temperature and Kinetic Energy')
ax1.set_xlabel('Time (ns)')
ax1.set_ylabel('Temperature (K)', color='g')
ax2.set_ylabel('Kinetic Energy (kJ/mol)', color='b')
plt.savefig('Temperature and kinetic energy',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()
```
