```python
import matplotlib.pyplot as plt
import numpy as np
import os

data = np.loadtxt("energy.xvg",skiprows=24)
print(data)

plt.hist(data[:,1],bins=100)
plt.xlabel('Potential Energy (kJ/mol)')
plt.ylabel('Count')
plt.title('Potential Energy Histogram at 300 K')
plt.savefig('Potential Energy Histogram at 300 K.pdf',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()
```
