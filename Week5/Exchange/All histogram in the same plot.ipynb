```python
import matplotlib.pyplot as plt
import numpy as np
import os

data_300 = np.loadtxt("T300/energy.xvg",skiprows=24)
print(data_300)


data_420 = np.loadtxt("T420/energy.xvg",skiprows=24)
print(data_420)

data_540 = np.loadtxt("T540/energy.xvg",skiprows=24)
print(data_540)

data_660 = np.loadtxt("T660/energy.xvg",skiprows=24)
print(data_660)

bins = np.linspace(-70, 70, 140)

plt.hist(data_300[:,1], bins, alpha=0.5, label='300 K')
plt.hist(data_420[:,1], bins, alpha=0.5, label='420 K')
plt.hist(data_540[:,1], bins, alpha=0.5, label='540 K')
plt.hist(data_660[:,1], bins, alpha=0.5, label='660 K')
plt.legend(loc='upper right')
plt.xlabel('Potential Energy (kJ/mol)')
plt.ylabel('Count')
plt.title('Potential Energy Histogram at different temperatures')
plt.savefig('Potential Energy Histogram at different temperatures.pdf',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()
# 


```
