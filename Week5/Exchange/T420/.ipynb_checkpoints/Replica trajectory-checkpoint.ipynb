```python
import matplotlib.pyplot as plt
import numpy as np
import os

data=np.loadtxt("replica_index.xvg",skiprows=0)
print(data)
```

```python
print(data[:,0])
```

```python
plt.plot(data[::1000,0], data[::1000,1])
plt.xlabel('Time (ps)')
plt.ylabel('Replica Index')
plt.title('Replica Exchange at 420K for every 100 ps')
plt.yticks(np.arange(0, max(data[::1000,1]+1), 1))
plt.ylim([0,3.2])
plt.xlim([0,5000])
plt.savefig('Replica Exchange at 420K for every 100 ps.pdf',format='pdf',bbox_inches='tight',pad_inches=0.1)
plt.show()
```
